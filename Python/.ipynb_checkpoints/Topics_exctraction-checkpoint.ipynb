# Extraction of topics from Wikipedia pages

```python
import numpy as np
import networkx as nx
import requests
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
### Importation of the Graph from gexf file ###

Graph = nx.read_gexf("peaks_graph_mod.gexf") 


### Creating Pandas DataFrame from Graph for convenient access ###

dataFrame = pd.DataFrame.from_dict(dict(Graph.nodes(data=True)), orient='index')
dataFrame['Qid'] = ''
dataFrame['topic'] = ''
dataFrame['probability'] = ''
dataFrame = dataFrame[['Qid', 'label', 'topic', 'probability', 'Modularity Class']]
print(dataFrame.info())
dataFrame
```

```python
### Extract nodes labels from Wikipedia queries ###


# Find the key of the value 'kv'  in a nested dictionnary
def findkeys(node, kv):
    if isinstance(node, list):
        for i in node:
            for x in findkeys(i, kv):
               yield x
    elif isinstance(node, dict):
        if kv in node:
            yield node[kv]
        for j in node.values():
            for x in findkeys(j, kv):
                yield x

            


# URL for the quieries
urls = "https://en.wikipedia.org/w/api.php?action=query&prop=pageprops&format=json&pageids="


i=0
Nb_pages = len(dataFrame)


### Associate the 'Qid' value of Wikipedia pages in the DataFrame ###
### Give '-1' value if an error has occured during the query      ###

for pageId in dataFrame.index:
    response = requests.get(urls + pageId).json()
    try:
        Qid = (list(findkeys(response, 'wikibase_item'))[0])
    except IndexError:
        dataFrame.loc[pageId, 'Qid'] = '-1'
        i+=1
        print(i,"/", Nb_pages, dataFrame.loc[pageId, 'label'], "   error")
    else:
        dataFrame.loc[pageId, 'Qid'] = Qid
        i+=1
        print(i,"/", Nb_pages, dataFrame.loc[pageId, 'label'])

        
# Save the DataFrame with their associated Qids
dataFrame.to_csv('pages_Qids.csv', encoding='utf-8')
```

```python
import numpy as np
import networkx as nx
import requests
import pandas as pd
import csv
import matplotlib.pyplot as plt
import time

### src countains Qid, Topic, Probability, Page Id, Wiki_DB ###
src = pd.read_csv('Match_topic_API.csv')

### dst countains modularity_class, clusters ##
dst = pd.read_csv('nodes.csv')




def change_modularity(src, dst):
    src['modularity_class'] = ''
    src['cluster'] = ''
    for page_index in dst['Id']:
        src.loc[src['page_id'] == page_index, 'modularity_class'] = dst.loc[dst['Id'] == page_index, 'modularity_class'].item()
        src.loc[src['page_id'] == page_index, 'cluster'] = dst.loc[dst['Id'] == page_index, 'Cluster'].item()
        src.loc[src['page_id'] == page_index, 'degree'] = dst.loc[dst['Id'] == page_index, 'Degree'].item()
    return src

changed_modularity = change_modularity(src, dst)

changed_modularity.to_csv('page_cluster.csv', encoding='utf-8')

changed_modularity
```

```python
import numpy as np
import networkx as nx
import requests
import pandas as pd
import csv
import matplotlib.pyplot as plt




def string_separator(df):
    for index, row in df.iterrows():
        topic_str = row['topic']
        topic_list = topic_str.split('.') 
        df.loc[df['topic'] == topic_str, 'subtopic'] = topic_list[-1]

        



dataFrame = pd.read_csv('page_cluster.csv')
dataFrame.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'], inplace = True )










def find_main_topics_clusters(cluster_name, dataFrame, modularity_df):
    
    dataFrame_without_biography = dataFrame.loc[dataFrame['topic'] != 'Culture.Biography.Biography*']

    topic_group = ((dataFrame_without_biography.groupby(['topic', cluster_name]))).count()


    total_cluster = ((dataFrame_without_biography.groupby([cluster_name]))).count()
    total_cluster = total_cluster['Qid'].to_frame()
    total_cluster.rename(columns={'Qid':'total'}, inplace=True)

    topic_group = topic_group['Qid'].to_frame().reset_index('topic')
    topic_group.rename(columns={'Qid':'count'}, inplace=True)
    topic_group.sort_values(by = [cluster_name, 'count'], inplace=True, ascending = [True, False])

    topic_group['total'] = total_cluster['total']
    topic_group['ratio'] = topic_group['count'] / topic_group['total']
    topic_group['subtopic'] = ''
    string_separator(topic_group)


#    topic_group = topic_group.loc[ (topic_group['total'] > 0) & (topic_group['ratio'] > 0.05) ]

    nb_topic = dataFrame[cluster_name].max() + 1
    for i in range(nb_topic):
        str_main_topics = topic_group.loc[i, 'subtopic'].values[0] + ' (' + str((topic_group.loc[i, 'ratio'].values[0]).round(decimals=2)) + ') '
        str_main_topics += '/ ' + topic_group.loc[i, 'subtopic'].values[1] + ' (' + str((topic_group.loc[i, 'ratio'].values[1]).round(decimals=2)) + ') '
        str_main_topics += '/ ' + topic_group.loc[i, 'subtopic'].values[2] + ' (' + str((topic_group.loc[i, 'ratio'].values[2]).round(decimals=2)) + ') '
        dataFrame.loc[dataFrame[cluster_name] == i, (cluster_name+'_topic')] = str_main_topics
        
        page_id_max_degree = dataFrame.loc[dataFrame[cluster_name] == i, 'degree'].idxmax()
        str_degree = dataFrame.loc[page_id_max_degree, 'page_title'] + ' (' + str(int(dataFrame.loc[page_id_max_degree, 'degree'])) + ')'
        
        mod = {'main_topics':str_main_topics, 'higher_degree_article':str_degree}
        modularity_df = modularity_df.append(mod, ignore_index=True)
        
    return modularity_df
cluster_name = 'modularity_class'   

modularity_df = pd.DataFrame({'main_topics':[], 'higher_degree_article':[]})


# dataFrame.loc[dataFrame[cluster_name] == 5, 'degree'].idxmax()
# dataFrame.loc[228]

modularity_df = find_main_topics_clusters(cluster_name, dataFrame, modularity_df) 
# cluster_name = 'cluster'      
# find_main_topics_clusters(cluster_name, dataFrame) 

modularity_df.index.rename('modularity_class', inplace = True)

modularity_df.to_csv('modularity_topics.csv', encoding='utf-8')

modularity_df

```
